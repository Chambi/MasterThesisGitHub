A cell in a notebook can contain text or code.

Strg+m+m

is used. LaTeX code inside text cells will be compiled, when it is contained in a single dollar or double dollar sign environment:
$$ \int f(x) dx. $$

The current cell can be left and is executed/ compiled when 

Shift + Enter

is pressed.

The text can be structured by headings, which are special text/ markdown cells:

# Plot a function and customize the plot

```python
# This cell contains python code

# First pylab is imported, which contains some core components of python for calculations and plotting:
from pylab import * 

# The x-values used in the following are 100 values from 0 to 10
x = linspace(0,10,100)
```

```python
plot(x,sin(x))
show()
```

```python
plot(x,sin(x), label = "Plot 1")
# a second function with a customized linestyle and color
plot(x,x, '--', color = 'red', label = "Plot 2")

# add titles:
xlabel("Title for x-axis")
ylabel(r"Titles with LaTeX code $\sum_i^n$")
title("the title", style='italic', color = 'white',
        bbox={'facecolor':'grey', 'alpha':0.5}) # or simply title('the title')

grid() # add a grid
legend() # add a legend

show()
```

```python
figure()

subplot(211)
plot(x,sin(x), label = "Plot 1")

subplot(212)
# a second function with a customized linestyle and color
plot(x,x, '--', color = 'red', label = "Plot 2")

# add titles:
xlabel("Large title for x-axis", fontsize = 25)
ylabel(r"Titles with LaTeX code $\sum_i^n$")
title("the title", style='italic', color = 'white',
        bbox={'facecolor':'grey', 'alpha':0.5}) # or simply title('the title')

# Labels can be switched of or changed:
xticks([0,5,10],["la", "le", "lu"])
yticks([])

grid() # add a grid
legend() # add a legend

tight_layout()

savefig("the_output.svg") # the output appears in the folder of the IPython Notebook
show()
```
