```python
import panna as p
from pylab import *
import numpy as np
from numpy import *
```

```python
def rms_it(data, axis=0):
    return sqrt(mean(data ** 2, axis))
```

```python
p.plot_all()
```

```python
K_phi = 30e-3 # mV/deg phase detector (PFD) constant
error = p.z[0]/K_phi
p.x[0] = p.x[0]+600e-6
```

```python


tside = [0.0, 2.46e-5, 4.8e-5, 7.1e-5, 9.2e-5, 113e-6, 133e-6, 152e-6, 171e-6, 189e-6, 207e-6, 225e-6, 242e-6, 259e-6, 275e-6, 290e-6, 307e-6, 322e-6, 337e-6, 352e-6, 366e-6, 380e-6, 394e-6, 409e-6, 422e-6, 436e-6, 449e-6, 462e-6, 475e-6, 487e-6, 500e-6, 513e-6, 525e-6, 538e-6, 550e-6, 563e-6, 575e-6, 588e-6, 601e-6, 629e-6, 690e-6]                                    
```

```python
print len(tside), 1/(tside[20]-tside[19])

figure()
subplot(211)

i = 0
while i < (len(tside)-1):
    #bar(tside[i], 5.0, abs(tside[i+1]-tside[i]), alpha = 0.3, facecolor = 'grey')
    axvspan(tside[i], tside[i+1], facecolor='grey', alpha=0.2)
    i += 2
    

p.grey_title("Resulting Signal during a Frequency Ramp")
p.colorcycle(3)
plot(p.x[0], [max(p.y[0])]*len(p.x[0]), '--', lw = 1.5, color = "grey", label = "max ="+str(round(max(p.y[0]),2))+" V")
plot(p.x[0], [min(p.y[0])]*len(p.x[0]), '--', lw = 1.5, color = "grey", label = "min ="+str(round(min(p.y[0]),2))+" V")
plot(p.x[0], p.a[0]*10, lw = 1.5, label = "References Mixed (x10)")
plot(p.x[0], p.y[0], lw = 1.5, label = "Resulting Signal")
leg = legend()
leg.get_frame().set_alpha(0.7)
ylabel("Amplitude [V]")
xlim(-50e-6, 800e-6)
p.scale_xaxis(subplot(211), 1e6)


subplot(212)

xlim(-50e-6, 800e-6)
p.grey_title("Error Signal during a Frequency Ramp")
p.colorcycle(3)
#
rms = p.rms(error)
#rms = 2
mean = array([average(error)]*len(error))

plot(p.x[0], error)
plot(p.x[0], error, lw = 1, label = r"$ \Delta \varphi (t)$")
plot(p.x[0], mean+rms, '--', lw = 2, color = "grey", label = r"$\langle \Delta \varphi \rangle=$"+str(round(rms[0]+mean[0],2))+r"$^\circ$")
#plot(p.x[0], mean-rms, '--', color = "grey")
#plot(p.x[0], mean, '--', color = "grey")
ylabel(r"Phase Error [$^\circ$]")
xlabel(r"Time [$\mu$s]")
leg = legend()
leg.get_frame().set_alpha(0.7)
grid()

p.scale_xaxis(subplot(212), 1e6)

tight_layout()
savefig("Frequency_Ramp.png")
show()
 
```

When the reference begins the frequency ramp the error signal rises up to about five times of its rms value. A zoom into the error signal shows that after 5 $\mu$s the optical signal follows the reference signal, so the error has been tracked out. The large error at the beginning of the sequency might be due to adjusting the $I$-part of the PID controller not to a high enough value. 

Nevertheless, one should notice, that the achieved rms phase stability of the optical to the DDS signal in this measurement is exceptionally good and the excursion is one the order of a tens of a degree.
