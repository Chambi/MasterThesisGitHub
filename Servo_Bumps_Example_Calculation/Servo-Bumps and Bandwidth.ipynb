# Servo-Bumps and Bandwidth

## Background (Master thesis section 2.4)

The phase locked loop can not react infinitely fast to a change in the reference signal, since all components have a limited modulation bandwidth and response time. A fast changing signal means a contribution of high frequencies in the Laplace transform.
In the case of a fast changing input the error $\Delta \phi(s)$ can not be adjusted immediately.

Mathematically $H(s)$ and $E(s)$ are determined by the dependence on the open loop gain $G(s)$.
Both have a pole at $1 + G(s) = 0$
, i.e. where the absolute value $|G(s)|=1$ and the argument $\angle(G(s)) = 180^\circ\ (e^{- i\pi} = -1)$.

Divergence of the transfer functions that describes the loop leads to instability of the system. As a measure of system stability the \textit{phase margin} is defined as the difference of $\angle(G(s))$ to $-180^\circ$ at the point of unity gain $|G(s)|=1$ \cite{Lunze2008, Ogata2001}.

For $|1+G(s)| < 1$ 
the system can still be stable but looses its ability to follow the reference properly. Therefore controlled oscillator phase and reference phase differ when this condition is met. 
	 Therefore, the system transfer function $H(s)$ will drop from unity, while the error transfer function rises $E(s)>0$ for large $s$. This is reminiscent of the behaviour of a low- and a high-pass filter when the cut-off frequency is approached.

Following the suggestions of \cite[eq. 23]{Prevedelli1995} the following form of $G(s)$ is used as an example
$$
G(s) = \frac{2\pi f_{\textrm{bw}}}{s} e^{ -sT_{\textrm{delay}}} \cdot K .
$$

$K$ is a linear gain constant, which could be for example an amplifier for the error voltage acting as a filter. Usually every element will have a finite bandwidth $f_{\textrm{bw}}$. Additionally a delay line can occur, e.g. a long cable. The delay results in an additional phase factor of $e^{-i 2\pi f T_{\textrm{delay}}} =  e^{ -sT_{\textrm{delay}}}$ of the signal.

Assumed is that the reference oscillator shows stable spectrum $S_{\phi_{ref}}$, in which phase noise is suppressed up to a high degree. This can be taken into account as rapidly decaying phase noise, described by large negative powers $\alpha = -2$ and $\alpha = -3$ in equation \ref{eq:Sphiansatz} (page \pageref{eq:Sphiansatz})
$$ S_{\phi_{ref}}(f) \propto f^{-2} + f^{-3} .$$

The phase locked loop aims to minimize the difference spectrum of $\Delta \phi(s)$ to the reference spectrum $S_{\phi_{ref}}(f)$. The connection between the reference spectrum and the difference of the oscillator to the reference is given by the error transfer function

$$ E(s) = \frac{\Delta\phi(s)}{\phi_{ref}(s)} \quad \Rightarrow S_{\Delta \phi}(f) = |E(2\pi i f)|^2 S_{\phi_{ref}}(f) .$$

The observed oscillator spectrum is $S_{\phi_{osc}}(f) = S_{\Delta \phi}(f) + S_{\phi_{ref}}(f)$. \\
Equation \ref{eq:GPrev} contains bandwidth $f_{\textrm{bw}}$ and delay time $T_{\textrm{delay}}$ as parameters. Arbitrary values for the bandwidth ($f_{\textrm{bw}} = 2.5$\,MHz) and for the delay time ($T_{\textrm{delay}} = 600$\,ns) are chosen. Now that the reference phase noise spectrum $S_{\phi_{ref}}(f)$ and the open-loop transfer function are given for this example system, its transfer functions can be calculated.

A plot of the resulting system transfer functions $H(s)$ and $E(s)$ reveals that their Bode plot looks reminiscent to the ones of a low and a high pass filter (figure \ref{fig:HE}).

```python
from pylab import * 
import panna as p 
import numpy as np
```

```python
frequency = arange(1.0,10.0e7,0.5e3)

# model for frequency noise: white + 1/f 
white = flicker = 1.0e5
S_f = lambda f: (white + flicker/f)

# open loop gain, tau: Grenzfrequenz
Tt = 600e-9
G = lambda f,f_grenz: 1.0/(1.j*f/f_grenz)*exp(-1j*pi*2*f*Tt)#*exp(-0.5*1.j*f/f_grenz*pi/2.0)

# error transfer function
Error = lambda f,K,f_grenz: 1.0/(1.0 + K*G(f,f_grenz))

# system transfer function
H = lambda f, K, f_grenz: K*G(f,f_grenz)/(1.0+K*G(f,f_grenz))
    
f_grenz = 2.5e6
K = array([0.001, 0.5, 1.0, 2.5])
Ks = array([0.001, 0.01, 0.05, 0.1, 0.15])#[0.001, 0.1, 0.15, 0.2])
```

## Open-Loop Transfer Function $G(s)$

```python
figure("G(s)")
Plot_abs = subplot(111) 
p.colorcycle(5)
for K in Ks:
    plot(frequency,10*log10(abs(K*G(frequency, f_grenz))), lw = 1.5, label=r'$|G(s)|$ for $K=$'+str(K))#$|G(2\pi i f)|$')
#plot(frequency,10*log10(abs(1-G(frequency, f_grenz))))
#plot(frequency, [0]*len(frequency), '--', color = 'grey', lw = 1.5, label = r'-180$^\circ$')

#plot(frequency, [0]*len(frequency), '--', color = 'grey', lw = 3)
xscale('log')
xlim(1e3, 3e6)
ylim(-40,40)
ylabel('Amplitude [dB]')
xlabel(r'Offset Frequency [Hz]')
grid(True, which = 'both')
legend()

Plot_phase = Plot_abs.twinx()
ylim(-200,-90)

Phase = np.unwrap(np.angle(G(frequency,f_grenz)))*180/pi
#plot(frequency,np.unwrap(np.angle(G(frequency, f_grenz)))
plot(frequency, Phase,'--', lw = 1.5, label=r'$arg(G(2\pi i f))$')
plot(frequency, [-180]*len(frequency), '--', color = 'grey', lw = 1.5, label = r'-180$^\circ$')
ylabel(r"Phase [$^\circ$]")


#grid(True, which = 'both', ls = '-')
legend(loc = 2)

savefig("Servo_Bumps_G.png")
show()
```

## Expectation for the spectral shape seen on a spectrum analyzer

```python
### Plot of the Power Spectral Density I ###

figure("PSD")

p.grey_title("Phase Noise Spectrum")

p.colorcycle(5)

subplot(111)
grid(True)

frequency = np.arange(-1e6,1e6,0.5e3)#1.0,5.0e6,0.5e3)
yscale('log')


plot(frequency, S_f(frequency)*1.0/frequency**2, lw = 3, color = 'grey', label = r'$ S_\phi\ ref  (f)$')

for K in Ks:
    #plot(frequency, abs(K*G(frequency,f_grenz)/(1.0 + K*G(frequency,f_grenz)))**2*S_f(frequency)*1.0/frequency**2, lw = 2, label = str(round(K,1)))
    plot(frequency, abs(1.0/(1.0 + K*G(frequency,f_grenz)))**2*S_f(frequency)*1.0/frequency**2 + S_f(frequency)*1/frequency**2, lw = 2, label = r'$K=$'+str(round(K,3)))
    #plot(frequency, abs(G(frequency, f_grenz))**2*S_f(frequency)/frequency**2, lw = 2, label = str(round(K,1)))
    
leg = plt.legend(loc=1, ncol=1, shadow=True, fancybox=True, numpoints=1)#, title = 'K value')
leg.get_frame().set_alpha(0.75)

#ylim(1e-9,1e0)
xlim(-650e3, 650e3)

xlabel(r'Frequency [Hz]')
ylabel(r'Amplitude [dB]')

savefig("Servo_Bumps_and_Bandwidth_PSD.png")

show()
```

## Low- and High-pass filter behaviour of the transfer functions $H(s)$ and $E(s)$

```python
### Plot of Transfer Functions ###
frequency = arange(1.0e2,3.0e6,0.5e3)

figure("Transfer Functions")

subplot(121)
p.grey_title(r'$|H(2\pi i f)|$')
p.colorcycle(5)
for K in Ks:
    plot(frequency, 10*log10(abs(H(frequency, K, f_grenz))), lw = 1.5, label = str(K))
xlabel(r'Frequency $f$ [Hz]')
ylabel(r'Amplitude [dB]')
grid(True, which = 'both')
xlim(frequency[0], frequency[-1])
xscale('log')
     
subplot(122)
p.grey_title(r'$|E(2\pi i f)|$')
p.colorcycle(5)
for K in Ks:
    plot(frequency, 10*log10(abs(Error(frequency, K, f_grenz))), lw = 1.5, label = r'$K=$'+str(round(K,3)))
xlabel(r'Frequency $f$ [Hz]')
ylabel(r'Amplitude [dB]')
grid(True, which = 'both')
xlim(frequency[0], frequency[-1])
xscale('log')
leg = plt.legend(loc=4, ncol=1, shadow=True, fancybox=True, numpoints=1)#, title = 'K value')
leg.get_frame().set_alpha(0.75)

tight_layout()
savefig("Servo_Bump_Transfer_Function.png")
show()
```
