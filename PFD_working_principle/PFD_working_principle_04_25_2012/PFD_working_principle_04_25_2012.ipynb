```python
import panna as p
from pylab import *

#p.plot_all()
data = p.load_all()
p.show_n()
len_files = p.number_of_files()
x,y,z,a,b = [0]*len_files, [0]*len_files, [0]*len_files, [0]*len_files, [0]*len_files

for i in arange(0, len_files):
    len_columns = p.number_of_columns(i)
    for j in arange(0, len_columns-1):
        if j == 0: x[i] = array(data[i][:,j])
        if j == 1: y[i] = array(data[i][:,j])
        if j == 2: z[i] = array(data[i][:,j])
        if j == 3: a[i] = array(data[i][:,j])
        if j == 4: b[i] = array(data[i][:,j])
```

```python
import numpy as np

def rising_edge(x,y):
    rising_edge_times = []
    dy = np.diff(y)
    minimum, maximum, height = min(y), max(y), max(y)-min(y)
    i = 0
    while i<len(dy)-10:
        x_possible = []
        while (np.average(dy[i:i+10])>0) and (y[i]>(minimum+0.2*height)) and (y[i]<(maximum-0.2*height)):
                x_possible.append(x[i])
                i += 1
        if (len(x_possible)>5): 
            rising_edge_times.append(np.average(x_possible))
        i+= 1
    #print rising_edge_times
    rising_edge_times.sort()
    return rising_edge_times

def PFD(time, LO_edges, RF_edges):
    i, lo, rf = 0, 0, 0
    UP, DOWN, TIME = [], [], []
    while i <= len(time):
        if lo >= len(LO_edges): break
        if rf >= len(RF_edges): break
        if (time[i] >= RF_edges[rf]):
            #print "RF edge",rf
            rf += 1
            while (time[i] <= LO_edges[lo]):
                if lo >= len(LO_edges): break
                if rf >= len(RF_edges): break
                if (time[i] >= RF_edges[lo]):
                    rf += 1
            # RF leads -> UP = 1 until LO rises
                UP.append(1.0)
                DOWN.append(0.0)
                TIME.append(time[i])
                i+= 1
            lo += 1
        if (time[i] >= LO_edges[lo]):
            lo += 1
            #print "LO edge",lo
            while (time[i] <= RF_edges[rf]):
                if lo >= len(LO_edges): break
                if rf >= len(RF_edges): break
                if (time[i] >= LO_edges[lo]):
                    lo += 1
                # RF lags -> DOWN = 1
                UP.append(0.0)
                DOWN.append(1.0)
                TIME.append(time[i])
                i += 1
            rf += 1
        else:
            # when the other one rises bring both back to 0 and next edge, no action if the one who rose before rises again
            UP.append(0.0)
            DOWN.append(0.0)
            TIME.append(time[i])
            i += 1
    return UP, DOWN, TIME
```

```python
# data in nanoseconds, no offset
offset = -1.061e6
factor = 1e9

for i in [5,7,9]:
    x[i] = x[i]*factor + offset
```

```python
# Find edges in comparator output
edges_RF = rising_edge(x[7], z[7])
edges_LO = rising_edge(x[7], a[7])
# Find Expectation for PFD output
up, down, time = PFD(x[7], edges_LO, edges_RF)
```
