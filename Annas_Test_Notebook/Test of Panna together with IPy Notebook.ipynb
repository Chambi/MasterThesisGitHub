# Test of Panna together with IP[y]: Notebook

Panny.py should search for *.csv files contained in the folder from which the Notebook was started. If the csv-files contain a header before the data begins, it will be skipped automatically. It will associate an index to each csv-file it found. The data in the columns of the file can then be accessed via this index:

```python
import panna as p
from pylab import *
from matplotlib import colors

data = p.load_all()
x, y, z = data[0][:,0], data[0][:,1], data[0][:,2]
```

Panny.py also contains functions to make the plot more beautiful (colorcycle and grey_title):

```python
figure()
p.colorcycle(3)
p.grey_title("lala")
plot(x,y)
plot(x,z)
xlabel(r"a proper x label $\Delta$ [unit]", color = '0.4')
grid()
show()
```
